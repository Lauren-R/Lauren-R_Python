## 2.

```python
# Find the mean number of orders by department
ords_prods_merge.groupby('department_id').agg({'order_number' : ['mean']})
```

## 3.

### These results include all departments instead of those only in the subset. The means are more accurate because they include all data instead of the limited subset of data.

## 4. 

```python
# 1. Split data into groups based on "user_id" column.
# 2. Apply the transform() function on the “order_number” column to generate the maximum orders for each user.
# 3. Create a new column, “max_order,” into which you’ll place the results of your aggregation.
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['customer_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
# Create flag to assign a "loyalty" label to a user ID based on corresponding max order value.
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge.head()
```

```python
# Check loyalty flag counts
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

## 5. 

```python
# Mean, min, max for loyalty flag groups
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

## 6. 

```python
# Create average price by customer column
ords_prods_merge['average_price'] = ords_prods_merge.groupby(['customer_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(100)
```

```python
# Create flag to assign a "average_spending" label to a user ID based on corresponding average product price value.
ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'average_spending'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price'] >= 10, 'average_spending'] = 'High spender'
```

```python
ords_prods_merge['average_spending'].value_counts(dropna = False)
```

## 7. 

```python
# Create column for average_days_since_prior_order by user
ords_prods_merge['average_days_since_prior_order'] = ords_prods_merge.groupby(['customer_id'])['days_since_prior_order'].transform(np.mean)
```

```python
ords_prods_merge.head(100)
```

```python
# Create flag to assign a "customer_frequency" label to a user ID based on corresponding average_days_since_prior_order column.
ords_prods_merge.loc[ords_prods_merge['average_days_since_prior_order'] > 20, 'customer_frequency'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['average_days_since_prior_order'] > 10) & (ords_prods_merge['average_days_since_prior_order'] <= 20), 'customer_frequency'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_days_since_prior_order'] <= 10, 'customer_frequency'] = 'Frequent customer'
```

```python
# Check customer_frequency flag counts
ords_prods_merge['customer_frequency'].value_counts(dropna = False)
```
